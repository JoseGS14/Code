```python
from scipy import stats

print("                   WELCOME TO OUR PROGRAM TO CALCULATE THE TRIMMED MEAN              ")
datas=[]    
i = 0
n = 0
g=0
porcentaje=0
m=0

e = float(input("Introduce the percentage to calculate:"))

for k in range(0,2):
    d = int(input("\nIntroduce the Datas:"))
    datas.append(d)
    
    porcentaje=e/100
    g=datas
    g.sort()
    print(f"\nThe correct order is: {g}")
    m = stats.trim_mean(g, porcentaje)
    print(f"\nThe Trimmed mean is: {m}")
    n += 1
    
while i < 1:
    
    print("\n                   ENTER 0 TO STOP THE PROCESS                     ")
    d = int(input("\nIntroduce the Datas:"))
    datas.append(d)
    
    
    porcentaje=e/100
    g=datas
    g.sort()
    print(f"\nThe correct order is: {g}")
    m = stats.trim_mean(g, porcentaje)
    print(f"\nThe Trimmed mean is: {m}")
    
    if d != 0:
        d = int(input("\nIntroduce the Datas:"))
        datas.append(d)
        
        porcentaje=e/100
        g=datas
        g.sort()
        print(f"\nThe correct order is: {g}")
        m = stats.trim_mean(g, porcentaje)
        print(f"\nThe Trimmed mean is: {m}")
        n += 1
        
    else:
        i += 1
```
